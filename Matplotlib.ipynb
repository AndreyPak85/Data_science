```python
import pandas as pd
from datetime import datetime

air_df = pd.read_csv('./data/air_quality.tar.gz', compression='gzip', sep=';', encoding='utf8')
air_df = air_df[air_df.iloc[:,0].notnull()]

# формируем 
air_df['ts'] = (air_df.iloc[:,0]+ ' ' + air_df.iloc[:,1]).map(
    lambda x: datetime.strptime(x.replace('.',':'), "%d/%m/%Y %H:%M:%S")
)

air_df.tail(30)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from pandas.plotting import register_matplotlib_converters
%matplotlib inline

register_matplotlib_converters()

fig, subplot = plt.subplots()
x_ticks = air_df['ts'].values[-30:-1]
y_ticks = air_df['NO2(GT)'].values[-30:-1]
subplot.plot(x_ticks, y_ticks)
plt.show()
```

```python
import numpy as np
from pandas.plotting import register_matplotlib_converters
%matplotlib inline

register_matplotlib_converters()

fig, subplot = plt.subplots()
x_values = air_df['T'].values[-30:-1]
y_values = air_df['NO2(GT)'].values[-30:-1]
subplot.scatter(
    x=x_values, y=y_values, 
    marker='o', c='red', edgecolor='blue'
)
```
